```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

new1Data50 = pd.read_csv('a1.csv',index_col=False)
new2Data50 = pd.read_csv('a2.csv',index_col=False)
# new3Data50 = pd.read_csv('a3.csv',index_col=False)
```

```python
plt.plot('Lz','Error(cpp)',data=new1Data50)
plt.grid()
plt.xlabel("Lz ($\AA$)")
plt.ylabel("Error (kcal/mol)")
plt.title("Variation of Error(cpp) in total energy vs Lz, BoxSize: 35 ($\AA$)")
```

```python
plt.plot('Lz','Error(cpp)',data=new2Data50)
plt.grid()
plt.xlabel("Lz ($\AA$)")
plt.ylabel("Error (kcal/mol)")
plt.title("Variation of Error(cpp) in total energy vs Lz, BoxSize: 35 ($\AA$)")
```

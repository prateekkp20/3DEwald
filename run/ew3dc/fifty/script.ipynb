```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

new1Data50 = pd.read_csv('a1.csv',index_col=False)
new1Data50['Error(cpp)'] = new1Data50['Error(cpp)'].apply(lambda x: x*4.184)
new1Data50['Error(lammps)'] = new1Data50['Error(lammps)'].apply(lambda x: x*4.184)
new2Data50 = pd.read_csv('a2.csv',index_col=False)
new2Data50['Error(cpp)'] = new2Data50['Error(cpp)'].apply(lambda x: x*4.184)
new2Data50['Error(lammps)'] = new2Data50['Error(lammps)'].apply(lambda x: x*4.184)
new3Data50 = pd.read_csv('a3.csv',index_col=False)
new3Data50['Error(cpp)'] = new3Data50['Error(cpp)'].apply(lambda x: x*4.184)
new3Data50['Error(lammps)'] = new3Data50['Error(lammps)'].apply(lambda x: x*4.184)
```

```python
# plt.plot('Lz','Error(cpp)',data=new1Data50)
plt.plot('Lz','Error(cpp)',data=new1Data50[new1Data50['Error(cpp)']<400])
plt.grid()
plt.xlabel("Lz ($\AA$)")
plt.ylabel("Error (kJ/mol)")
plt.title("Variation of Error(cpp) in total energy vs Lz, BoxSize: 50, 400 Atoms ($\AA$)")
```

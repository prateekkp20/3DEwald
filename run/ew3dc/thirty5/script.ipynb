```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

new1Data35 = pd.read_csv('a1.csv',index_col=False)
new1Data35['Error(cpp)'] = new1Data35['Error(cpp)'].apply(lambda x: x*4.184)
new1Data35['Error(lammps)'] = new1Data35['Error(lammps)'].apply(lambda x: x*4.184)
new2Data35 = pd.read_csv('a2.csv',index_col=False)
new2Data35['Error(cpp)'] = new2Data35['Error(cpp)'].apply(lambda x: x*4.184)
new2Data35['Error(lammps)'] = new2Data35['Error(lammps)'].apply(lambda x: x*4.184)
# new3Data50 = pd.read_csv('a3.csv',index_col=False)
```

```python
plt.plot('Lz','Error(cpp)',data=new1Data35)
plt.grid()
plt.xlabel("Lz ($\AA$)")
plt.ylabel("Error (kJ/mol)")
plt.title("Variation of Error(cpp) in total energy vs Lz, BoxSize: 35 ($\AA$)")
```

```python
plt.plot('Lz','Error(cpp)',data=new2Data35)
plt.grid()
plt.xlabel("Lz ($\AA$)")
plt.ylabel("Error (kJ/mol)")
plt.title("Variation of Error(cpp) in total energy vs Lz, BoxSize: 35 ($\AA$)")
```
